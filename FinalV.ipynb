```
class Rational:

  def __init__(self, numerator,denumerator = 1):
    self.numerator = numerator
    self.denumerator = denumerator

  def __str__(self):
    return (str(self.numerator)+ "/" + str(self.denumerator))

  def normalize(self):
    if(self.denumerator<0):
      self.denumerator = self.denumerator*-1
    self.numerator = self.numerator+self.denumerator
    self.denumerator = self.numerator - self.denumerator
    self.numerator = self.numerator - self.denumerator

  def __add__(self, other):
    return Rational(self.numerator*other.denumerator+other.numerator*self.denumerator,self.denumerator * other.denumerator)

  def __sub__(self, other):
    return Rational(self.numerator*other.denumerator-other.numerator*self.denumerator,self.denumerator * other.denumerator)

  def __mul__(self, other):
    return Rational(self.numerator*other.numerator,self.denumerator * other.denumerator)

  def __truediv__(self,other):
    return  Rational(self.numerator*other.denumerator,self.denumerator * other.numerator)

  def __pow__(self,other):
    return Rational(self.numerator**other,self.denumerator**other)
  def __radd__(self, other):
    return Rational(self.numerator*other.denumerator+other.numerator*self.denumerator,self.denumerator * other.denumerator)

  def __rsub__(self, other):
    return Rational(self.numerator*other.denumerator-other.numerator*self.denumerator,self.denumerator * other.denumerator)

  def __rmul__(self, other):
    return Rational(self.numerator*other.numerator,self.denumerator * other.denumerator)

  def __rtruediv__(self,other):
    return Rational(self.numerator*other.denumerator,self.denumerator * other.numerator)
  def __neg__(self):
    return Rational(self.numerator-self.denumerator,self.denumerator)
  def __abs__(self):
    return Rational(abs(self.numerator),abs(self.denumerator))
  def __iadd__(self, other):
    self.numerator = self.numerator*other.denumerator+other.numerator*self.denumerator
    self.denumerator = self.denumerator * other.denumerator
  def __isub__(self, other):
    self.numerator = self.numerator*other.denumerator-other.numerator*self.denumerator
    self.denumerator = self.denumerator * other.denumerator
  def __imul__(self, other):
    self.numerator = self.numerator*other.numerator
    self.denumerator = self.denumerator*other.denumerator
  def __idiv__(self, other):
    self.numerator = self.numerator*other.denumerator
    self.denumerator = self.denumerator*other.numerator
  def __lt__(self,other):
    n1 = self.numerator*other.denumerator
    n2 = other.numerator*self.denumerator
    if n1<n2:
      return True
    else:
      return False
  def __le__(self,other):
    n1 = self.numerator*other.denumerator
    n2 = other.numerator*self.denumerator
    if n1<=n2:
      return True
    else:
      return False
  def __gt__(self,other):
    n1 = self.numerator*other.denumerator
    n2 = other.numerator*self.denumerator
    if n1>n2:
      return True
    else:
      return False
  def __ge__(self,other):
    n1 = self.numerator*other.denumerator
    n2 = other.numerator*self.denumerator
    if n1>=n2:
      return True
    else:
      return False
  def __eq__(self,other):
    if self.numerator*other.denumerator == other.numerator*self.denumerator:
      return True
    else:
      return False
  def __ne__(self,other):
    if self.numerator*other.denumerator != other.numerator*self.denumerator:
      return True
    else:
      return False
  def evalf(self):
    return self.numerator/self.denumerator
  def cut(self):
    d = self.numerator//self.denumerator
    n = d*self.denumerator-self.numerator
    return d and Rational(n,self.denumerator)
```

```
r1 = Rational(6,3)
print(r1)
r1.normalize()
print("r1: "+str(r1))
r2 = Rational(1,2)
print("r2: "+str(r2))
print("+: "+str(r1+r2))
print("-: "+str(r1-r2))
print("/: "+str(r1/r2))
print("*: "+str(r1*r2))
print("^2: "+str(r1**2))
print("==: "+str(r1==r2))
print("!=: "+str(r1!=r2))
print("<: "+str(r1<r2))
print("<=: "+str(r1<=r2))
print(">: "+str(r1>r2))
print(">=: "+str(r1>=r2))
print("evalf: "+str(r1.evalf()))
print("cut: "+str(r1.cut()))
```
